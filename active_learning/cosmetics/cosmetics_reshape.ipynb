```python
import pandas as pd
```

## Starting from summary

```python
results = pd.read_csv('all_data_summary.csv')
results.head()
```

```python
df = results.copy()
```

```python
df['year_month_event_type'] = df['year_month'] + '_' + df['event_type']
df.head()
```

```python
melted_df = pd.melt(df, id_vars=['user_id', 'year_month_event_type'], 
    value_vars=['product_id_nunique', 'price_sum', 'price_mean', 'price_median', 'user_session_nunique', 'category_id_nunique'])

melted_df['joined'] = melted_df['year_month_event_type'] + '_' + melted_df['variable']
```

```python
melted_df.head()
```

```python
pivoted_df = pd.pivot(melted_df, index=['user_id'], columns='joined', values='value')

# Print the resulting pivoted DataFrame
pivoted_df = pivoted_df.reset_index()
pivoted_df.head()
```

```python
pivoted_df.columns.to_list()
```

```python
pivoted_df.to_csv('organized_data.csv', index=False)
```
