## Loading parameters

```python
with open("hyperparameter_search.json", "r") as file:
    experiments = json.load(file)["experiments"]
```

```python
for exp in experiments:
    print(f"Running experiment with learning_rate={exp['learning_rate']} and batch_size={exp['batch_size']}")
```
