```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
def simulateBeacons(location, nbeacons, freq, totaltime, meanRssi, sdRssi):
  #totaltime = 20
  #freq = 100

  tsamples1b=(1000/freq)*totaltime
  tsamples=tsamples1b*nbeacons
  res_signal= np.random.normal(meanRssi, sdRssi, int(tsamples))

  #area=[location]*int(tsamples)

  a=list(range(1,nbeacons+1))
  nameBeacs=[location + str(i) for i in a]
  res_nameBeacs =  [ele for ele in nameBeacs for i in range(int(tsamples1b))]

  tmpDF = pd.DataFrame([res_nameBeacs,res_signal]).T
  tmpDF.columns = ['name', 'rssi']
  tmpDF['timestamp'] = tmpDF.groupby('name')['name'].rank(method='first')
  return(tmpDF)

finalDF = simulateBeacons('study', nbeacons=2, freq=100, totaltime=40, meanRssi=-65, sdRssi=5)
finalDF.head()
```

```python
def expKitchen(experimentNumber):
  user_at='kitchen'
  experiment=experimentNumber

  tmpk = simulateBeacons('kitchen', nbeacons=2, freq=100, totaltime=40, meanRssi=-52, sdRssi=4.8)
  tmpl = simulateBeacons('living', nbeacons=2, freq=100, totaltime=40, meanRssi=-74, sdRssi=4.5)
  tmpm = simulateBeacons('main', nbeacons=2, freq=100, totaltime=40, meanRssi=-84, sdRssi=4.5)
  tmps = simulateBeacons('study', nbeacons=2, freq=100, totaltime=40, meanRssi=-70, sdRssi=4.3)

  finalK = tmpk.append(tmpl, ignore_index=True)
  finalK = finalK.append(tmpl, ignore_index=True)
  finalK = finalK.append(tmpm, ignore_index=True)
  finalK = finalK.append(tmps, ignore_index=True)
  finalK['experiment'] = experiment
  finalK['user_at'] = user_at
  return(finalK)
```
